2. Предложите способ заменить отсутствующие значения из пункта 1 и реализуйте его. Способ нужно аргументировать(сказать почему это достаточно хорошее приближение)

### Анализ

По графикам видны выбросы с очень большими площадями,
а также выбросы с жилой площадью больше общей. Избавляемся от них.

Используем медианное значение отношение жилой площади к общей из отфильтрованных данных, это позволяет заполнить пропуски и не исказить общие данные. Медианное значение в данном случае лучше, так как оно меньше реагирует на выбросы и позволяет получить лучшее приближение.

(такой проблемы нет в исходных данных, но в другой ситуации могло пригодится)

На случай, если неизвестны обе площади, пытаемся заполнить медианным значением для конкретного района и количества комнат. Это позволит заполнить пропуски на основе информации о районе, где большая вероятность типовой застройки с одинаковыми параметрами квартир.

Для оставшихся данных используем медиану всех жилых площадей.


````python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))

valid_data = df[df['life_sq'].notnull() & df['full_sq'].notnull()]

plt.subplot(2, 2, 1)
plt.scatter(valid_data['full_sq'], valid_data['life_sq'], alpha=0.5)
plt.xlabel('Общая площадь (full_sq)')
plt.ylabel('Жилая площадь (life_sq)')
plt.title('Корреляция: full_sq и life_sq')

# есть выбросы с очень большими значениями, убираем их
full_sq_upper = valid_data['full_sq'].quantile(0.99)
life_sq_upper = valid_data['life_sq'].quantile(0.99)

valid_data = valid_data[
    (valid_data['full_sq'] <= full_sq_upper) & 
    (valid_data['life_sq'] <= life_sq_upper)
]

plt.subplot(2, 2, 2)
plt.scatter(valid_data['full_sq'], valid_data['life_sq'], alpha=0.5)
plt.xlabel('Общая площадь (full_sq)')
plt.ylabel('Жилая площадь (life_sq)')
plt.title('Корреляция без выбросов: full_sq и life_sq')

plt.subplot(2, 2, 3)
# через clip убираем выбросы, где жилая площадь больше общей
ratio = (valid_data['life_sq'] / valid_data['full_sq']).clip(0, 1)
plt.hist(ratio, bins=100, alpha=0.7)
plt.xlabel('Отношение life_sq / full_sq')
plt.ylabel('Частота')
plt.title('Распределение отношения жилой к общей площади')

plt.tight_layout()
plt.show()

print('Статистика отношения life_sq / full_sq:')
print(f'Среднее: {ratio.mean():.3f}')
print(f'Медиана: {ratio.median():.3f}')
print(f'Стандартное отклонение: {ratio.std():.3f}')
````

### Заполняем данные

````python
df_filled = df.copy()
median_ratio = ratio.median()

mask = df_filled['life_sq'].isnull() & df_filled['full_sq'].notnull()
df_filled.loc[mask, 'life_sq'] = (df_filled.loc[mask, 'full_sq'] * median_ratio).round()

if df_filled['life_sq'].isnull().any():
    group_median = df_filled.groupby(['sub_area', 'num_room'])['life_sq'].transform('median')
    df_filled['life_sq'] = df_filled['life_sq'].fillna(group_median)

if df_filled['life_sq'].isnull().any():
        overall_median = df_filled['life_sq'].median()
        df_filled['life_sq'] = df_filled['life_sq'].fillna(overall_median)

if df_filled['life_sq'].notnull().all():
      print('Все значения заполнены!')

df_filled['life_sq'] = df_filled['life_sq'].astype('Int64')
````

Проверяем результат заполнения: медиана сохранилась на прежнем месте, а среднее значение изменилось менее чем на 2%. Можно сделать вывод, что заполнение получилось с достаточно хорошим приближением. 

````python
plt.figure(figsize=(12, 10))

valid_data = df_filled[df_filled['life_sq'].notnull() & df_filled['full_sq'].notnull()]

plt.subplot(2, 2, 1)
plt.scatter(valid_data['full_sq'], valid_data['life_sq'], alpha=0.5)
plt.xlabel('Общая площадь (full_sq)')
plt.ylabel('Жилая площадь (life_sq)')
plt.title('Корреляция: full_sq и life_sq')

full_sq_upper = valid_data['full_sq'].quantile(0.99)
life_sq_upper = valid_data['life_sq'].quantile(0.99)

valid_data = valid_data[
    (valid_data['full_sq'] <= full_sq_upper) & 
    (valid_data['life_sq'] <= life_sq_upper)
]

plt.subplot(2, 2, 2)
plt.scatter(valid_data['full_sq'], valid_data['life_sq'], alpha=0.5)
plt.xlabel('Общая площадь (full_sq)')
plt.ylabel('Жилая площадь (life_sq)')
plt.title('Корреляция без выбросов: full_sq и life_sq')

plt.subplot(2, 2, 3)
# через clip убираем выбросы, где жилая площадь больше общей
ratio = (valid_data['life_sq'] / valid_data['full_sq']).clip(0, 1)
plt.hist(ratio, bins=100, alpha=0.7)
plt.xlabel('Отношение life_sq / full_sq')
plt.ylabel('Частота')
plt.title('Распределение отношения жилой к общей площади')

plt.tight_layout()
plt.show()

print('Статистика отношения life_sq / full_sq:')
print(f'Среднее: {ratio.mean():.3f}')
print(f'Медиана: {ratio.median():.3f}')
print(f'Стандартное отклонение: {ratio.std():.3f}')
````

9. Проанализируйте есть ли разница в цене между квартирами на 1 и 9 этажах

````python
floor_1 = df[df['floor'] == 1]
floor_9 = df[df['floor'] == 9]
floor_1_price = floor_1['price_doc']
floor_9_price = floor_9['price_doc']

mean_1 = floor_1_price.mean()
mean_9 = floor_9_price.mean()
median_1 = floor_1_price.median()
median_9 = floor_9_price.median()

floor_1_price_per_m = (floor_1_price / floor_1['full_sq']).median()
floor_9_price_per_m = (floor_9_price/ floor_9['full_sq']).median()

m_1 = floor_1['full_sq'].median()
m_9 = floor_9['full_sq'].median()

print('Этаж      Среднее (руб.)   Медиана (руб.)    Стоимость м^2 (руб.)   Медианная площадь (м)')
print(f'  1        {mean_1:10,.0f}       {median_1:10,.0f}       {floor_1_price_per_m:10,.0f}\t\t{m_1:10,.0f}')    
print(f'  9        {mean_9:10,.0f}       {median_9:10,.0f}       {floor_9_price_per_m:10,.0f}\t\t{m_9:10,.0f}')
print()
print(f'Разница средней стоимости: {mean_9 - mean_1:+,.0f} руб. ({((mean_9/mean_1)-1)*100:+.1f}%)')
print(f'Разница медианной стоимости: {median_9 - median_1:+,.0f} руб. ({((median_9/median_1)-1)*100:+.1f}%)')
print(f'Разница медианной стоимости кв.м: {floor_9_price_per_m - floor_1_price_per_m:+,.0f} руб. ({((floor_9_price_per_m/floor_1_price_per_m)-1)*100:+.1f}%)')
print(f'Разница медианной площади: {m_9 - m_1:+,.0f} м ({((m_9/m_1)-1)*100:+.1f}%)')
````

Квартиры на 9 этаже на 15-20% дороже, что полностью объясняется их большей площадью (+9%). При этом стоимость квадратного метра остается практически одинаковой на обоих этажах. Можно утвержать, что стоимость квадратного метра не зависит от выбора между 1 и 9 этажами, однако квартиры на 9 этаже в целом дороже из-за большей площади.
